```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("C:\\Users\\soura\\Downloads\\householdtask3.csv")

# Group data by year and calculate mean values for the required columns
grouped_data = data.groupby('year').mean()

# Bar Chart: Distribution of total households (`tot_hhs`) across years
plt.figure(figsize=(12, 6))
plt.bar(grouped_data.index, grouped_data['tot_hhs'], color='skyblue', label='Total Households')
plt.title('Total Households Distribution Across Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Households', fontsize=12)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Line Chart: Comparing household income and expenditure across years
plt.figure(figsize=(12, 6))
plt.plot(grouped_data.index, grouped_data['income'], marker='o', label='Income', color='green')
plt.plot(grouped_data.index, grouped_data['expenditure'], marker='s', label='Expenditure', color='orange')
plt.title('Household Income and Expenditure Across Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```
